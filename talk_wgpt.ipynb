```python
import os
import openai
from pydub import AudioSegment
from pydub.playback import play
from gtts import gTTS
import speech_recognition as sr


def speak(text): 
    
    tts = gTTS(text=text, lang='en')

    filename = "audio/speech"
    tts.save(filename + '.mp3')
    
    sound = AudioSegment.from_mp3(filename + '.mp3')
    sound.export(filename + '.wav', format="wav")
    song = AudioSegment.from_wav(filename + '.wav')
    
    play(song)
    
    os.remove(filename + '.mp3')
    os.remove(filename + '.wav')

    
def gpt(prompt):
    
    openai.api_key = "OPENAI API KEY"

    response = openai.Completion.create(
      model="text-davinci-003",
      prompt=prompt,
      temperature=0.0,
      max_tokens=200,
      top_p = 1.0,
      frequency_penalty=0.0,
      presence_penalty=0.0
    )

    return response['choices'][0]['text'].strip()
    
    
def main():
    
    r = sr.Recognizer()

    while(True):   

        try:

            with sr.Microphone() as source2: #Device can be added manually as an argument, e.g. device_index=1

                r.adjust_for_ambient_noise(source2, duration=0.2)
                audio2 = r.listen(source2)

                MyText = r.recognize_google(audio2)
                MyText = MyText.lower()
                print(MyText+'\n')
                
                if len(MyText) > 0:
                    response = gpt(MyText)
                    print(response+'\n')
                    speak(response)
            

        except sr.RequestError as e:
            print("Could not request results; {0}".format(e))

        except sr.UnknownValueError:
            print("unknown error occurred")
            
            
main()
```
